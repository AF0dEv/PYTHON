```python
# O Pandas é a melhor Biblioteca Python para Dados
# O Pandas é a melhor Biblioteca Python para Análise de Dados
```

```python
# A Primeira coisa a saber é a Utilização do import
import pandas as pd # normalizaçao da comunidade
```

```python
# O Pandas funciona dentro do Python como um DataFrame
# Um DataFrame é uma Tabela dentro do Python
```

```python
# Existem 3 Formas de Criar uma DataFrame
```

```python
# 1ª Forma --> Criar uma DataFrame Vazio
dataframe = pd.DataFrame()
```

```python
# 2ª Forma --> Criar um DataFrame a partir de um Dicionário
venda = {'data': ['15/12/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['batata', 'arroz'],
         'qtd': [50, 70],
        }
print(venda)
```

```python
# Agora Criar DataFrame a partir do Dicionário
vendas_df = pd.DataFrame(venda)
print(vendas_df)
```

```python
display(vendas_df)
```

```python
# 3ª Forma --> Criar uma DataFrame a partir de um Ficheiro ou Base de Dados
vendas_df = pd.read_excel('Vendass.xlsx')
display(vendas_df)
```

```python
# Existem 3 Métodos muito Uteis:
# -> Head
# -> Shape
# -> Describe
```

```python
# Método HEAD
# Por defeito mostra 5 PRIMEIRAS LINHAS
# Se quisermos mais, temos de Colocar na Função do Método 
display(vendas_df.head(10))
```

```python
# Método SHAPE
# Serve para Mostrar quantas Linhas e Colunas o DataFrame tem
print(vendas_df.shape)
```

```python
# Método DESCRIBE
# Mostra um Resumo dos Dados (apenas Dados NUMÉRICOS) da DataFrame
# Por TOTAL, MÉDIA, DESVIO PADRÃO, MIN, MAX, ETC..
display(vendas_df.describe())
```
